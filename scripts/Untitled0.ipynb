```python
# Leer la hoja correcta del Excel

df = pd.read_excel("Dataset_Ignes.xlsm", sheet_name="Base Normalizada")

# Mostrar primeras filas
df.head()
```

```python
# Información general del dataset
print("Cantidad de filas y columnas:")
print(df.shape)

print("\nNombres de columnas:")
print(df.columns)

print("\nTipos de datos:")
print(df.dtypes)
```

```python
# Estadísticas básicas
df.describe()
```

```python
# Ver valores faltantes
df.isnull().sum()
```

```python
# Promedio de precios
print("Promedio de precios:")
print(df["Precio (Moneda)"].mean())
```

```python
# Cantidad de autos por tipo
print(df["Tipo"].value_counts())
```

```python
import matplotlib.pyplot as plt

# Gráfico de tipos de autos
df["Tipo"].value_counts().plot(kind="bar")

plt.title("Cantidad de autos por tipo")
plt.xlabel("Tipo")
plt.ylabel("Cantidad")
plt.show()
```

```python
# Eliminar columnas innecesarias
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Ver columnas finales
print(df.columns)
```

```python
# Eliminar columnas duplicadas
df = df[['Año', 'Tipo', 'Kilometraje', 'Motor',
         'Transmision', 'Combustible',
         'Propietarios', 'Precio (Moneda)', 'Vendedor']]

# Ver dataset limpio
df.head()
```

```python
# Conclusiones automáticas del análisis

print("CONCLUSIONES DEL DATASET")
print("------------------------")

print("Cantidad total de registros:", len(df))

print("Precio promedio:",
      round(df["Precio (Moneda)"].mean(), 2))

print("Precio máximo:",
      df["Precio (Moneda)"].max())

print("Tipo de vehículo más frecuente:")
print(df["Tipo"].mode()[0])
```

```python
# Guardar dataset limpio en CSV

df.to_csv("dataset_limpio.csv", index=False)

print("Archivo CSV guardado correctamente")
```

```python
# Guardar conclusiones en archivo TXT

with open("conclusiones.txt", "w", encoding="utf-8") as archivo:
    archivo.write("CONCLUSIONES DEL DATASET\n")
    archivo.write("------------------------\n")
    archivo.write(f"Cantidad total de registros: {len(df)}\n")
    archivo.write(f"Precio promedio: {round(df['Precio (Moneda)'].mean(),2)}\n")
    archivo.write(f"Precio máximo: {df['Precio (Moneda)'].max()}\n")
    archivo.write(f"Tipo más frecuente: {df['Tipo'].mode()[0]}\n")

print("Archivo TXT guardado correctamente")
```
